# Functions, Fibonaccis

```python
def fib1(n):
    ''' Calculates the fib sequence as a dict '''
    a = [x+1 for x in range(n)]
    f1=0
    f2=1
    f=[f1, f2]
    while len(f) < n:
        fn = f[-1] + f[-2]
        f.append(fn)
    fib_dict = dict(zip(a,f))
    return fib_dict
print(fib1(10))

def fib2(n):
    ''' Calculates the fib sequence as a dict '''
    a = 0 
    b = 1
    d = dict()
    for i in range(1,n+1):
       d[i] = a
       a,b = b,a+b
    return d
print(fib2(10))

def fib3(n):
    ''' Calculates the nth fib '''
    a = 0
    b = 1
    for i in range(n-1):
        a,b = b,a+b
    return a
print(fib3(10))
```

# Classes

## BankAccount

```python
class BankAccount(object):
    
    def __init__(self, balance=0.0):
        self.balance = balance
        
    def display_balance(self):
        print(self.balance)
        
    def deposit(self, amount):
        self.balance += amount
        print(self.balance)
        
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(self.balance)
        else:
            print('insufficient balance')

a1 = BankAccount()
a1.display_balance()
a1.deposit(10)
a1.withdraw(20)
a1.withdraw(5)
```

# Card

```python
class Card(object):

    suits = {'h':'hearts','d':'diamonds','c':'clubs','s':'spades'}
    named = {'a':'ace','j':'jack','q':'queen','k':'king'}
    
    def __init__ (self, value, suit):
        self.value = Card.named.get(value) if value.isalpha() else value
        self.suit = Card.suits.get(suit)

    def get_value(self):
        return self.value

    def get_suit(self):
        return self.suit

    def __str__(self):
        card = str(self.value) + ' of ' + str(self.suit)
        return card

x = Card('a','s')
print(x)
```

# Brackets test

## Without stacks

```python
# brackets = { 40:41, 123:125, 91:93 }
# if brackets.get(ord(test[i]), 0) != ord(test[length-i-1]):

# 422ms

def brackets_test(test):
    brackets = {'(':')','[':']','{':'}'}
    print(test, end=': ')
    match = True
    length = len(test)
    if length % 2:
        match = False
    else:
        for i in range(length//2):
            if match:
                if brackets.get(test[i], 0) != test[length-i-1]:
                    match = False
    print('Match') if match else print('Fail')
    print()

test1 = '[{()}]' # good
test2 = '[{(}]' # bad, odd
test3 = '[{(})]' # bad, unmatched
test4 = '{{([{()}])}}' # good, longer
test5 = '}{([})}}' # bad, start
tests = [test1, test2, test3, test4, test5]

for test in tests:
    brackets_test(test)
```

## With stacks

```python
# 390ms

def balanced_brackets(s): 
    stack = []
    brackets = {'(':')','[':']','{':'}'}
    for char in s:
        if char in brackets.keys():
            stack.append(char)
        else:
            if len(stack) == 0 or brackets[stack.pop()] != char:
                return False
    return len(stack) == 0

test1 = '[{()}]' # good
test2 = '[{(}]' # bad, odd
test3 = '[{(})]' # bad, unmatched
test4 = '{{([{()}])}}' # good, longer
test5 = '}{([})}}' # bad, start
tests = [test1, test2, test3, test4, test5]

for test in tests:
    brackets_test(test)
```

# Sorting

```python
def insertion_sort(l):
    n = len(l)
    for i in range (1,n):
        x = l[i]
        j = i
        # print(f'\n{x} in position {i+1}')
        # print(l)
        while j > 0 and l[j-1] > x:
            l[j] = l[j-1]
            j = j-1
        l[j] = x
    return l
test = [45,87,6,70,2,5,3,24]
print(test)
insertion_sort(test)
```
