# Your First Machine Learning Model

```python
home_data.columns
```

```python
y = home_data.SalePrice

feature_names = [
    'LotArea',
    'YearBuilt',
    '1stFlrSF',
    '2ndFlrSF',
    'FullBath',
    'BedroomAbvGr',
    'TotRmsAbvGrd'
]

X = home_data[feature_names]
```

```python
iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(X, y)
```

```python
predictions = iowa_model.predict(X)
```

```python
home_data['Predictions'] = predictions
home_data.head(5).sort_values(by='LotArea')
```

# Model Validation

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(train_X, train_y)
```

```python
print("Sale prices")
print(y[:5].tolist())

print("")

print("In-sample predictions")
print(iowa_model.predict(X.head()))

print("")

print("Out-of-sample predictions")
val_predictions = iowa_model.predict(val_X)
print(val_predictions[:5])
```

```python
mae = mean_absolute_error(y, predictions)
print(f"In-sample MEA: {int(mae)}")

print("")

val_mae = mean_absolute_error(val_y, val_predictions)
print("Out-of-sample MAE: {:,.0f}".format(val_mae))
```

# Underfitting and Overfitting

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]

for n in candidate_max_leaf_nodes:
    mae = get_mae(n, train_X, val_X, train_y, val_y)
    print(f"Nodes: {n} \t MAE: {int(mae)}")
```

```python
final_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=0)
final_model.fit(X, y)
```

# Random Forests

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X, train_y)
iowa_rf_predictions = rf_model.predict(val_X)
print(f"Validation MAE for random forest model: {int(mean_absolute_error(val_y, iowa_rf_predictions))}")
```
